```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:/Internship/raw data/student/student-mat.csv', sep=';')

```

```python
print(df.head())
```

```python
## checking if the  data is null
missing_values = df.isnull().sum()
print("The missing values ing the column sre:\n", missing_values)
```

```python
## column data type
print ("\nData types of column:\n", df.dtypes)
```

```python
## database size

print("\n Shape of the database is:", df.shape)
```

```python
## data cleaning
df.drop_duplicates(inplace=True)

df.dropna(inplace=True) 
```

```python
print(df)
```

```python
df.columns.tolist()
```

```python
## calculating the average math score

avg_math_score = df['G3'].mean()

print("The avg math score(G3) is:" , avg_math_score)
```

```python
## students scored above 15 in their final grade (G3)

students_above_15 = (df['G3']> 15).sum()

print("The students who scored above 15 is:",students_above_15)
```

```python
## gender with higher final avg grade

avg_female_g3 = df[df['sex'] == 'F']['G3'].mean()
print("Female: ",avg_female_g3)

avg_male_g3 = df[df['sex'] == 'M']['G3'].mean()
print("Male: ",avg_male_g3)

if avg_male_g3 > avg_female_g3 :
    print("Males have a higher avg final grade")
elif avg_female_g3 > avg_male_g3:
    print("Females have a higher avg final grade")
else:
    print("Both have same final grade")
```

```python
## visual representation

## Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['G3'], bins=10, kde=True) 
plt.xlabel('Final Grade (G3)')
plt.ylabel('Number of Students')
plt.title('Distribution of Final Grades')
plt.show()
```

```python
## Scatter plot between Study Time and Final Grade

plt.figure(figsize=(8,6))
sns.scatterplot(x= 'studytime', y='G3', data=df)
plt.xlabel('Study Time')
plt.ylabel('Final Grade(G3)')
plt.title('Studty Time vs. Final Grade')
plt.show()
```

```python
## Bar Chart Comparison
avg_scores = df.groupby('sex')['G3'].mean()
plt.figure(figsize=(6, 4))
sns.barplot(x=avg_scores.index, y=avg_scores.values)
plt.xlabel('Gender')
plt.ylabel('Average Final Grade (G3)')
plt.title('Average Scores by Gender')
plt.show()
```
